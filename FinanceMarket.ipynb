```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import the dataset
data = pd.read_csv("F:/bagher/Market.csv")
```

```python
#making a dataframe and calling it
df = pd.DataFrame(data)
df
```

```python
#sorting the data by 'NYA' index
df1 = df[df['Index'] == 'NYA']
pd.concat([df1.head(5), df1.tail(5)])
```

```python
#removing the decimal part of the numbers
df2 = np.around(df1)
pd.concat([df2.head(5), df2.tail(5)])
```

```python
#removing the 'Volume' column
df3 = df2.drop(['Volume'], axis = 1)
df3
```

```python
#Adj Close base on Open plot with noises
plt.figure(figsize = (14, 6))
plt.scatter(df3["Open"],df3["Adj Close"] , color = 'brown')
plt.xlabel('Open', fontsize = 18)
plt.ylabel('Adj Close', fontsize = 18)
plt.grid()
```

```python
#finding noise 1
n1 = df3[(
    df3['Adj Close'] > 6000) & (df3['Open'] < 1000)]
n1
```

```python
##fiing noise 2
n2 = df3[(df3['Adj Close'] < 1000) &  (df3['Open'] > 5000)]
n2
```

```python
#remiving noises
df4 = df3.drop(index = [831, 852])
df4
```

```python
#Adj Close base on Open plot without noises
plt.figure(figsize = (14, 6))
plt.scatter(df4["Open"],df4["Adj Close"] , color = 'red')
plt.xlabel('Open', fontsize = 18)
plt.ylabel('Adj Close', fontsize = 18)
plt.grid()
plt.savefig("F:/bagher/plot.png")
```

```python
#Adj Close base on High plot with noises
plt.figure(figsize = (14, 6))
plt.scatter(df4["High"],df4["Adj Close"] , color = 'darkgreen')
plt.xlabel('High', fontsize = 18)
plt.ylabel('Adj Close', fontsize = 18)
plt.grid()
plt.savefig("F:/bagher/plot.png")
```

```python
#finding noise 3 and 4
n3 = df4[(df4['Adj Close'] < 1000) & (df4['High'] > 6000)]
n3
```

```python
#removing noises
df5 = df4.drop(index = [829, 833])
df5
```

```python
#Adj Close base on High plot without noises
plt.figure(figsize = (14, 6))
plt.scatter(df5["High"],df5["Adj Close"] , color = 'lime')
plt.xlabel('High', fontsize = 18)
plt.ylabel('Adj Close', fontsize = 18)
plt.grid()
```

```python
plt.figure(figsize = (14, 6))
plt.scatter(df5["Low"],df5["Adj Close"] , color = 'dimgrey')
plt.xlabel('Low', fontsize = 18)
plt.ylabel('Adj Close', fontsize = 18)
plt.grid()
#a normal plot and there is no noise to find
```

```python
plt.figure(figsize = (14, 6))
plt.scatter(df5["Close"],df5["Adj Close"] , color = 'rebeccapurple')
plt.xlabel('Close', fontsize = 18)
plt.ylabel('Adj Close', fontsize = 18)
plt.grid()
#a normal plot and there is no noise to find
```

```python
#detecting missing values
df5.isnull().sum()
```

```python
#removing missing values
df6 = df5.dropna()
```

```python
df6
```

```python
#making a new dataframe nad correcting the date format for time series plots
df7 = df6.copy()
df7['Date'] = pd.to_datetime(df7['Date']).dt.date
```

```python
pd.concat([df7.head(5), df7.tail(5)])
```

```python
#Adj Close base on Date plot
f = plt.figure()
f.set_figwidth(24)
f.set_figheight(10)
plt.title("Adj Close base on Date", fontsize = 26)
plt.plot(df7['Date'], df7['Adj Close'])
plt.xlabel("Date", fontsize = 23)
plt.ylabel("Adj Close", fontsize = 23)
plt.grid()
plt.show()
```
